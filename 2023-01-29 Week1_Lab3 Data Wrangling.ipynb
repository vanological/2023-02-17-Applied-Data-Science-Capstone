```python
landing_outcomes = df['Outcome'].value_counts()
landing_outcomes
```

```python
landing_outcomes_dict = {}
for i, key in enumerate(landing_outcomes.index):
    landing_outcomes_dict[key] = i
print(landing_outcomes_dict)
```

```python
bad_outcomes = set(landing_outcomes.keys()[[1,3,5,6,7]])
bad_outcomes
```

```python
LandingOutcome = []
for i in df['Outcome']:
    if (i in bad_outcomes):
        LandingOutcome.append(0)
    else:
        LandingOutcome.append(1)
print(LandingOutcome)
df['Class'] = LandingOutcome
```

```python
df.drop('landing_class', axis=1, inplace=True)
```
