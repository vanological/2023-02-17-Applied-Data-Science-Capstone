```python
data['date_utc'] = data['date_utc'].map(lambda x: datetime.datetime.strptime(x, '%Y-%m-%dT%H:%M:%S.%fZ').date())
```

```python
data.rename({'date_utc': 'date'}, axis=1, inplace=True)
```

```python
data.head(2)
```

```python
data = data[data['date'] <= datetime.date(2020, 11, 13)]
```

```python
BoosterVersion = []
PayloadMass = []
Orbit = []
LaunchSite = []
Outcome = []
Flights = []
GridFins = []
Reused = []
Legs = []
LandingPad = []
Block = []
ReusedCount = []
Serial = []
Longitude = []
Latitude = []
```

```python
counter = 0
print('total query # {}'.format(len(data['rocket'])))
for rocket in data['rocket']:
#    print(type(rocket))
    response = requests.get('https://api.spacexdata.com/v4/rockets/'+ rocket)
    if response.status_code == 200:
        response = response.json()
        counter += 1
        BoosterVersion.append(response['name'])
print('Total queried {}'.format(counter))
        
```

```python
len(BoosterVersion)
```

```python
print('total query # {}'.format(data['payloads'].shape))
counter = 0
for payload in data['payloads']:
    response = requests.get('https://api.spacexdata.com/v4/payloads/'+payload)
    if response.status_code == 200:
        counter += 1
        PayloadMass.append(response.json()['mass_kg'])
        Orbit.append(response.json()['orbit'])
        #print('total queried {}'.format(counter))
print(len(PayloadMass), len(Orbit))
```

```python
# Takes the dataset and uses the launchpad column to call the API and append the data to the list
def getLaunchSite(data):
    for x in data['launchpad']:
       if x:
         response = requests.get("https://api.spacexdata.com/v4/launchpads/"+str(x)).json()
         Longitude.append(response['longitude'])
         Latitude.append(response['latitude'])
         LaunchSite.append(response['name'])

# Takes the dataset and uses the cores column to call the API and append the data to the lists
def getCoreData(data):
    for core in data['cores']:
            if core['core'] != None:
                response = requests.get("https://api.spacexdata.com/v4/cores/"+core['core']).json()
                Block.append(response['block'])
                ReusedCount.append(response['reuse_count'])
                Serial.append(response['serial'])
            else:
                Block.append(None)
                ReusedCount.append(None)
                Serial.append(None)
            Outcome.append(str(core['landing_success'])+' '+str(core['landing_type']))
            Flights.append(core['flight'])
            GridFins.append(core['gridfins'])
            Reused.append(core['reused'])
            Legs.append(core['legs'])
            LandingPad.append(core['landpad'])
```

```python
getLaunchSite(data)
print(len(Longitude))
```

```python
getCoreData(data)
print(len(Outcome))
```
