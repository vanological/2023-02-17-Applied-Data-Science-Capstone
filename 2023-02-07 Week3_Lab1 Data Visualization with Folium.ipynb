```python
import folium
import wget
import pandas as pd
```

```python
from folium.plugins import MarkerCluster
from folium.plugins import MousePosition
from folium.features import DivIcon
```

```python
spacex_csv_file = wget.download('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv')
spacex_df=pd.read_csv(spacex_csv_file)
```

```python
spacex_df.head()
```

```python
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

```python
nasa_coordinate = (29.559684888503615, -95.0830971930759)
site_map = folium.Map(location=nasa_coordinate, zoom_start=10)
circle = folium.Circle(nasa_coordinate, radius=1000, fill=True, color='#d35400').add_child(folium.Popup(('#####')))
marker = folium.Marker(
    nasa_coordinate,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
    )
)
site_map.add_child(circle)
site_map.add_child(marker)
```

```python
import re
siteDict = {}
for index, col in launch_sites_df.iterrows():
    name=col.values[0]
    name=re.sub(r'-', '', name)
    name=re.sub(r'\s', '_', name)
    siteDict[name]=(col.values[1], col.values[2])
siteDict
```

```python
for key, value in siteDict.items():
    print('adding {} {}'.format(key, value))
    circleTrial = folium.Circle(value, radius=5000, color='#d35400', fill=True).add_child(folium.Popup(str(key)))
    markerTrial = folium.Marker(
        value,
        icon=DivIcon(
            icon_size=(10,10),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % key,
        ))
    site_map.add_child(circleTrial)
    site_map.add_child(markerTrial)
site_map
```

```python
ccaf_lc_coor = [28.562302, -80.577356]
ccaf_lc_map = folium.Map(ccaf_lc_coor, zoom_start=1)
ccaf_lc_circle = folium.Circle(ccaf_lc_coor, radius=1000, fill=True, color='#d35400').add_child(folium.Popup("CCAF LC-40"))
ccaf_lc_marker = folium.Marker(
    ccaf_lc_coor,
    icon=DivIcon(
        icon_size=(10,10),
        icon_anchor=(0,0),
        html='<div style="font-size:12; color:#d35400"><b>%s</b></div>' % 'CCAF LC40</div>',
        )
)
ccaf_lc_map.add_child(ccaf_lc_circle)
ccaf_lc_map.add_child(ccaf_lc_marker)
ccaf_lc_map.add_child(circle)
ccaf_lc_map.add_child(marker)
ccaf_lc_map
```

```python
marker_cluster = MarkerCluster()
```

```python
colorList=[]
for cell in spacex_df['class']:
    if cell == 0:
        colorList.append('red')
    else:
        colorList.append('green')
spacex_df['marker_color'] = colorList
spacex_df.tail()
```

```python
site_map.add_child(marker_cluster)

for index, record in spacex_df.iterrows():
    marker = folium.Marker(
        [record['Lat'], record['Long']],
        popup=index,
        icon=folium.Icon(color='white', icon_color=record['marker_color'],
        ))
    marker_cluster.add_child(marker)

site_map
```
